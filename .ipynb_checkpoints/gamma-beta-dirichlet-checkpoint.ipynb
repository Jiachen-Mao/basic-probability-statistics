## Gamma function

$\Gamma (\alpha) = \int_0^\infty t^{\alpha - 1} e^{-t} dt$, $\alpha > 0$

$\Gamma (\alpha + 1) = \alpha \Gamma (\alpha)$, $\Gamma (n+1) = n!$, $\Gamma (1) = 1$, $\Gamma (1/2) = \sqrt{\pi}$

```python
import numpy as np
from scipy.special import gamma
import matplotlib.pyplot as plt
import pylab
```

```python
fig = plt.figure(figsize = (8, 6))
x = np.linspace(-5, 5, 1000)
plt.plot(x, gamma(x), ls = '-', c = 'k', label = '$\Gamma(x)$')

x2 = np.linspace(1, 6, 6)
y = np.array([1, 1, 2, 6, 24, 120])
pylab.plot(x2, y, marker = '*', markersize = 12, ls = '', c = 'r', label = '$(x-1)!$')

plt.title('Gamma function')
plt.ylim(-50, 50)
plt.xlim(-5, 5)
plt.xlabel('$x$')
plt.legend()
plt.show()
```

## Gamma distribution

$f(x; \alpha, \beta) = \frac{\beta^\alpha x^{\alpha - 1} e^{-\beta x}}{\Gamma (\alpha)}$

```python
from scipy.stats import gamma
```

```python
alpha_values = [1, 2, 3, 3, 3]
beta_values = [0.5, 0.5, 0.5, 1, 2]
color = ['b', 'r', 'g', 'y', 'm']
x = np.linspace(1E-6, 10, 1000)

fig, ax = plt.subplots(figsize = (8, 6))

for k, t, c in zip(alpha_values, beta_values, color):
    dist = gamma(k, 0, t)
    plt.plot(x, dist.pdf(x), c = c, label = r'$\alpha = %.1f, \theta = %.1f$'% (k, t))

plt.xlim(0, 10)
plt.ylim(0, 2)
plt.xlabel('$x$')
plt.ylabel(r'$p(x;\alpha , \beta)$')
plt.title('Gamma distribution')
plt.legend(loc = 0)
plt.show()
```

## Beta distribution

$f(x) = \frac{\Gamma (\alpha + \beta)}{\Gamma (\alpha) \Gamma (\beta)} x^{\alpha - 1} (1 - x)^{\beta - 1}$

```python
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt
```

```python
alpha_values = [1/3, 2/3, 1, 1, 2, 20]
beta_values = [1, 2/3, 3, 1, 1, 20]
color = ['b', 'r', 'g', 'y', 'm', 'tab:pink']
x = np.linspace(0, 1, 1002)[1:-1]

fig, ax = plt.subplots(figsize = (10, 8))

for a, b, c in zip(alpha_values, beta_values, color):
    dist = beta(a, b)
    plt.plot(x, dist.pdf(x), c = c, label = r'$\alpha = %.1f, \beta = %.1f$'% (a, b))

plt.xlim(0, 1)
plt.ylim(0, 6)
plt.xlabel('$x$')
plt.ylabel(r'$p(x;\alpha,\beta)$')
plt.title('Beta distribution')

ax.annotate('Beta(1/3, 1)', xy = (0.014, 5), xytext = (0.04, 5.2), 
            arrowprops = dict(facecolor = 'black', arrowstyle = '-'))

ax.annotate('Beta(2/3, 2/3)', xy = (0.99, 2.4), xytext = (0.86, 2.8), 
            arrowprops = dict(facecolor = 'black', arrowstyle = '-'))

ax.annotate('Beta(1, 3)', xy = (0.06, 2.6), xytext = (0.07, 3.1), 
            arrowprops = dict(facecolor = 'black', arrowstyle = '-'))

ax.annotate('Beta(1, 1)', xy = (0.8, 1), xytext = (0.7, 2), 
            arrowprops = dict(facecolor = 'black', arrowstyle = '-'))

ax.annotate('Beta(2, 1)', xy = (0.9, 1.8), xytext = (0.75, 2.6), 
            arrowprops = dict(facecolor = 'black', arrowstyle = '-'))

ax.annotate('Beta(20, 20)', xy = (0.5, 5), xytext = (0.52, 5.4), 
            arrowprops = dict(facecolor = 'black', arrowstyle = '-'))

plt.show()
```

## Dirichlet distribution

$P(\textbf{X}; \alpha) = \frac{\Gamma (\sum_i \alpha_i)}{\Pi_i \Gamma (\alpha_i)} \Pi_i x_j^{\alpha_j - 1}$

```python
import matplotlib.tri as tri
```

```python
corners = np.array([[0, 0], [1, 0], [0.5, 0.75**0.5]])
triangle = tri.Triangulation(corners[:, 0], corners[:, 1])
midpoints = [(corners[(i + 1) % 3] + corners[(i + 2) % 3]) / 2.0 for i in range(3)]

def xy2bc(xy, tol=1.e-3):
    s = [(corners[i] - midpoints[i]).dot(xy - midpoints[i]) / 0.75 for i in range(3)]
    return np.clip(s, tol, 1.0 - tol)
```

```python
class Dirichlet(object):
    
    def __init__(self, alpha):
        from math import gamma
        from operator import mul
        from functools import reduce
        self._alpha = np.array(alpha)
        self._coef = gamma(np.sum(self._alpha)) / reduce(mul, [gamma(a) for a in self._alpha])
    
    def pdf(self, x):
        from operator import mul
        from functools import reduce
        return self._coef * reduce(mul, [xx ** (aa - 1) for (xx, aa) in zip(x, self._alpha)])
```

```python
def draw_pdf_contours(dist, nlevels = 200, subdiv = 8, **kwargs):
    import math

    refiner = tri.UniformTriRefiner(triangle)
    trimesh = refiner.refine_triangulation(subdiv = subdiv)
    pvals = [dist.pdf(xy2bc(xy)) for xy in zip(trimesh.x, trimesh.y)]

    plt.tricontourf(trimesh, pvals, nlevels, **kwargs)
    plt.axis('equal')
    plt.xlim(0, 1)
    plt.ylim(0, 0.75**0.5)
    plt.axis('off')
```

```python
draw_pdf_contours(Dirichlet([1, 1, 1]))
```

```python
draw_pdf_contours(Dirichlet([0.9, 0.9, 0.9]))
```

```python
draw_pdf_contours(Dirichlet([5, 5, 5]))
```

```python
draw_pdf_contours(Dirichlet([1, 2, 3]))
```
