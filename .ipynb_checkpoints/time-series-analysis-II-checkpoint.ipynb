# Time Series Analysis II

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.stattools import adfuller as ADF
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.stats.diagnostic import acorr_ljungbox
from statsmodels.tsa.arima_model import ARIMA
%matplotlib inline
%pylab inline
```

```python
# Display chinese in Jupyter

plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
pylab.rcParams['figure.figsize'] = (10, 6)
sns.set(color_codes = True)
```

## Import data

```python
data = pd.read_excel('arima_data.xls', index_col = u'Date')
data.plot()
```

## Time series analysis

```python
plot_acf(data).show()
```

```python
print(u'ADF on the original series: ', ADF(data[u'Sales']))
```

```python
# First-order analysis

D_data = data.diff().dropna()
D_data.columns = [u'SalesDiff']
D_data.plot()
```

```python
plot_acf(D_data).show()
```

```python
print(u'ADF on the first-order diff series: ', ADF(D_data[u'SalesDiff']))
```

```python
print(u'White noise test on the first-order diff series: ', acorr_ljungbox(D_data, lags = 1))
```

```python
plot_pacf(D_data).show()
```

```python
# Find the "best" model

data[u'Sales'] = data[u'Sales'].astype(float)

pmax = int(len(D_data)/10)
qmax = int(len(D_data)/10)
bic_matrix = []

for p in range(pmax + 1):
    tmp = []
    for q in range(qmax + 1):
        try:
            tmp.append(ARIMA(data, (p, 1, q)).fit().bic)
        except:
            tmp.append(None)
    bic_matrix.append(tmp)
    
bic_matrix = pd.DataFrame(bic_matrix)

p, q = bic_matrix.stack().idxmin()
print(u'p and q with the minimal BIC: %s, %s' %(p, q))
```

```python
model = ARIMA(data, (p, 1, q)).fit()
model.summary2()
```

```python
model.forecast(5)
```
