```r
x.rhodium.stdev = sqrt(var(x.rhodium))
x.rhodium.mean.sterr = x.rhodium.stdev / sqrt(length(x.rhodium))
print(x.rhodium.mean.sterr)
```

```r
alpha = 0.1
z.upperalphahalf = qnorm(1 - alpha/2)
x.rhodium.mean = mean(x.rhodium)
x.rhodium.mean.ci <- c(-1, 1)*z.upperalphahalf*x.rhodium.mean.sterr + x.rhodium.mean
print(x.rhodium.mean.ci)
```

```r
stats1.rhodium = c(x.rhodium.mean, x.rhodium.mean.ci)
```

```r
x.iridium.stdev = sqrt(var(x.iridium))
x.iridium.mean.sterr = x.iridium.stdev / sqrt(length(x.iridium))
print(x.iridium.mean.sterr)
```

```r
alpha = 0.1
z.upperalphahalf = qnorm(1 - alpha/2)
x.iridium.mean = mean(x.iridium)
x.iridium.mean.ci <- c(-1, 1)*z.upperalphahalf*x.iridium.mean.sterr + x.iridium.mean
print(x.iridium.mean.ci)
```

```r
stats1.iridium = c(x.iridium.mean, x.iridium.mean.ci)
```

```r
fcn.trimmedmean <- function(x, d, trim = 0){
    return(mean(x[d], trim/length(x)))
}

fcn.median <- function(x, d){
    return(median(x[d]))
}
```

```r
# Bootstrap analysis of sample 10% trimmed mean

x = x.rhodium
x.boot.trimmedmean.1 = boot(x, fcn.trimmedmean, trim = .1, R = 1000)
par(mfol = c(1, 2))
plot(x.boot.trimmedmean.1)
title(paste("Trimmed mean: StDEV = ", as.character(round(sqrt(var(x.boot.trimmedmean.1$t)), digits = 4)), sep = ""),
      adj = 1)
```

```r
print(x.boot.trimmedmean.1$t0)
```

```r
print(sqrt(var(x.boot.trimmedmean.1$t)))
```

```r
stats2.trim10.rhodium <- c(x.boot.trimmedmean.1$t0, (sqrt(var(x.boot.trimmedmean.1$t))))
boot.ci(x.boot.trimmedmean.1, conf = .9, type = "basic")
```

```r
# Bootstrap analysis of sample median

x = x.rhodium
x.boot.median = boot(x, fcn.median, R = 1000)
par(mfol = c(1, 2))
plot(x.boot.median)
title(paste("Median: StDEV = ", as.character(round(sqrt(var(x.boot.median$t)), digits = 4)), sep = ""),
      adj = 1)
```

```r
print(x.boot.median$t0)
```

```r
print(sqrt(var(x.boot.median$t)))
```

```r
boot.ci(x.boot.median, conf = .9, type = "basic")
```

```r
stats2.median.rhodium <- c(x.boot.median$t0, (sqrt(var(x.boot.median$t))))
stats1.rhodium
```

## Case study 2

```r
flowocc = read.table(file = "Rice 3e Datasets/ASCII Comma/Chapter 10/flow-occ.txt",
                   sep = ",", stringsAsFactors = FALSE, header = TRUE)

Timestamp2 = strptime(flowocc$Timestamp, "%m/%d/%Y %H:%M:%S")
flowocc$Timestamp = Timestamp2
```

```r
par(mfcol = c(2, 1))
plot(flowocc$Timestamp, flowocc[,"Lane.1.Occ"], main = "Lane 1: Occupancy", ylab = "", xlab = "")
plot(flowocc$Timestamp, flowocc[,"Lane.1.Flow"], main = "Lane 1: Flow", ylab = "", xlab = "")
```

```r
par(mfcol = c(1, 1))
boxplot(flowocc[,c("Lane.1.Flow", "Lane.2.Flow", "Lane.3.Flow")], ylab = "Count")
```

```r
par(mfcol = c(1, 1))
pairs(flowocc[,c("Lane.1.Flow", "Lane.2.Flow", "Lane.3.Flow")])
```

```r
par(mfcol = c(1, 1))
plot(flowocc$Lane.1.Flow, flowocc$Lane.2.Flow, xlab = "Lane 1", ylab = "Lane 2", main = 'Flow')
abline(a = 0, b = 1, col = 'red', lwd = 2)
```

```r
occ.mean = c(mean(flowocc$Lane.1.Occ), mean(flowocc$Lane.2.Occ), mean(flowocc$Lane.3.Occ))
occ.median = c(median(flowocc$Lane.1.Occ), median(flowocc$Lane.2.Occ), median(flowocc$Lane.3.Occ))

print(cbind(occ.mean, occ.median))
```

```r
par(mfcol = c(2, 2))
hist(flowocc$Lane.1.Occ, nclass = 25)
hist(flowocc$Lane.1.Occ, nclass = 50)
hist(flowocc$Lane.1.Occ, nclass = 100)
hist(flowocc$Lane.1.Occ, nclass = 200)
```

```r
par(mfcol = c(1, 3))

plot(flowocc$Lane.1.Occ, flowocc$Lane.2.Occ, xlab = "Lane 1 Occupancy", ylab = "Lane 2 Occupancy")
abline(a = 0, b = 1, col = 'red', lwd = 2)

plot(flowocc$Lane.1.Occ, flowocc$Lane.3.Occ, xlab = "Lane 1 Occupancy", ylab = "Lane 3 Occupancy")
abline(a = 0, b = 1, col = 'red', lwd = 2)

plot(flowocc$Lane.2.Occ, flowocc$Lane.3.Occ, xlab = "Lane 2 Occupancy", ylab = "Lane 3 Occupancy")
abline(a = 0, b = 1, col = 'red', lwd = 2)
```
