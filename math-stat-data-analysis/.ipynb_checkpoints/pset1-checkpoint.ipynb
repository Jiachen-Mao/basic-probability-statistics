```r
X = rnorm(1000)
Y = rnorm(1000)
W = X^2 + Y^2
V = atan(Y/X)

par(mfcol = c(2, 2))
hist(X, breaks = 100)
hist(Y, breaks = 100)
hist(W, breaks = 100)
hist(V, breaks = 100)
```

```r
set.seed(1)
wvec = rf(10000, df1 = 3, df2 = 3)
wvec.cummeans = cumsum(wvec)/c(1:length(wvec))
plot(wvec.cummeans)
abline(h = 3/(3-2))
```
