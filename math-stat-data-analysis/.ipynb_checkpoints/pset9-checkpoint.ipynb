## Regression analysis implemented in R

```r
# Problem 14.9.39

tankvolume = read.table(file = "Rice 3e Datasets/ASCII Comma/Chapter 14/tankvolume.txt",
                      sep = ",", stringsAsFactors = FALSE, header = TRUE)

Volume = tankvolume$Volume
Pressure = tankvolume$Pressure
```

```r
plot(Volume, Pressure)
summary(Volume)
```

```r
lmfit1 = lm(Pressure~ Volume)
summary(lmfit1)
```

```r
plot(Volume, Pressure)
abline(lmfit1, col = 'green')
```

```r
plot(Volume, lmfit1$residuals)
```

```r
VolumeSq = Volume * Volume

lmfit2 = lm(Pressure~ Volume + VolumeSq)
summary(lmfit2)
```

```r
plot(Volume, lmfit2$residuals)
abline(h = 0, col = 'gray')
```

```r
# Problem 10.9.50

flowocc = read.table(file = "Rice 3e Datasets/ASCII Comma/Chapter 10/flow-occ.txt",
                      sep = ",", stringsAsFactors = FALSE, header = TRUE)

Timestamp2 = strptime(flowocc$Timestamp, "%m/%d/%Y %H:%M:%S")
flowocc$Timestamp = Timestamp2
```

```r
x = flowocc$Lane.1.Occ
y = flowocc$Lane.3.Occ
plot(x, y, xlab = "Lane.1.Occ", ylab = "Lane.3.Occ")
```

```r
# Local linear regression

plot(x, y, xlab = "Lane.1.Occ", ylab = "Lane.3.Occ")

ksmooth.05 <- ksmooth(x, y, kernel = "normal", bandwidth = .05)
lines(ksmooth.05$x, ksmooth.05$y, col = 'red')

ksmooth.1 <- ksmooth(x, y, kernel = "normal", bandwidth = .1)
lines(ksmooth.1$x, ksmooth.1$y, col = 'green')

ksmooth.2 <- ksmooth(x, y, kernel = "normal", bandwidth = .2)
lines(ksmooth.2$x, ksmooth.2$y, col = 'blue')

ksmooth.4 <- ksmooth(x, y, kernel = "normal", bandwidth = .4)
lines(ksmooth.4$x, ksmooth.4$y, col = 'orange')

legend(x = 0., y = .5, legend = c("Bandwidth = .05", "Bandwidth = .10", "Bandwidth = .20", "Bandwidth = .40"),
      lty = c(1, 1, 1, 1), col = c('red', 'green', 'blue', 'orange'))
title(main = "Occupancy: Lane 3 vs Lane 1\n Local linear fits (normal kernel)")
```

```r
lmfit1 = lm(Lane.3.Occ ~ Lane.1.Occ, data = flowocc, x = TRUE, y = TRUE)
head(lmfit1$x)
```

```r
plot(flowocc$Lane.1.Occ, flowocc$Lane.3.Occ)

lmfit1 = lm(Lane.3.Occ ~ Lane.1.Occ, data = flowocc)
abline(lmfit1, col = 'green')
```

```r
plot(flowocc$Lane.1.Occ, lmfit1$residuals)
abline(h = 0, col = 'gray')
```

```r
qqnorm(lmfit1$residuals)
```

```r
oldpar = par(np.readonly = TRUE)
layout(matrix(c(1, 2, 3, 4), 2, 2))
plot(lmfit1)
```

```r
# Fit regression for two subsets

ind.subset1 = (flowocc$Lane.1.Occ < .18)
ind.subset2 = (flowocc$Lane.1.Occ > .18)

plot(flowocc$Lane.1.Occ[ind.subset1], flowocc$Lane.3.Occ[ind.subset1])
lmfit1.subset1 = lm(Lane.3.Occ ~ Lane.1.Occ, data = flowocc, weight = 1*ind.subset1)
abline(lmfit1.subset1, col = 'green')
```

```r
print(summary(lmfit1.subset1))
```

```r
plot(flowocc$Lane.1.Occ[ind.subset1], lmfit1.subset1$residuals[ind.subset1])

abline(h = 0, col = 'gray')
```

```r
qqnorm(lmfit1.subset1$residuals[ind.subset1])
```

```r
oldpar = par(np.readonly = TRUE)
layout(matrix(c(1, 2, 3, 4), 2, 2))
plot(lmfit1.subset1)
```

```r
# Fit one model with different regression for two subsets

lmfit.both = lm(flowocc$Lane.3.Occ ~ -1 +
               I(1*ind.subset1) + I(ind.subset1*flowocc$Lane.1.Occ) + 
               I(1*ind.subset2) + I(ind.subset2*flowocc$Lane.1.Occ),
               x = TRUE, y = TRUE)

options(width = 120)

head(lmfit.both$x)
```

```r
summary(lmfit.both)
```
